```python
panjang = float(input("Masukkan panjang: "))
lebar = float(input("Masukkan lebar: "))

luas = panjang * lebar
print(f"Luas segi empat ialah: {luas}")
```

```python
nombor = int(input("Masukkan satu nombor: "))

if nombor % 2 == 0:
    print(f"{nombor} ialah nombor GENAP.")
else:
    print(f"{nombor} ialah nombor GANJIL.")
```
