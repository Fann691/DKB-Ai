```python
import numpy as np

markah = np.array ([70,85,90,65,80])
print (markah)
```

KIRA MARKAH PURATA

```python
purata = np.mean (markah)
print ("purata markah")
```

```python
tertinggi = np.max (markah)
print ("tertinggi markah:",tertinggi)
```

```python
terendah = np.min (markah)
print ("terendah markah:",terendah)
```

JUMLAH

```python
jumlah = np.sum (markah)
print ("jumlah markah:",jumlah)
```

TAMBAH

```python
markah_baru = markah + 5
print ("markah baru:",markah_baru)
```

AMALI PANDAS

```python

import pandas as pd

data = {
    "Nama": ["ali","ameer","alim","kasim"],
    "Markah" : [70,85,90,65]
}


```

BINA DATAFRAME

```python
df = pd.DataFrame(data)
print(df)
```

purata = df["Markah"].mean()
print("Purata markah:",purata)

```python
df["status"] = np.where(df["Markah"] < 50, "cemerlang", "lulus")
print(df)
```

```python
def tentukan_gred(markah):
    if markah >= 80:
        return "A"
    elif markah >= 70:
        return "B"
    elif markah >= 60:
        return "C"
    elif markah >= 50:
        return "D"
    else:
        return "E"


df["Grade"] = df["Markah"].apply(tentukan_gred)
print(df)
```
